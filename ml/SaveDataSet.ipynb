```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
```

```python
link_source = pd.read_excel('D:\\ProgPrj\\dsProjects\\ЦП_ЮФО_Подготовка\\RealDataSet\\ссылки_исходные.xlsx')
link_new = pd.read_excel('D:\\ProgPrj\\dsProjects\\ЦП_ЮФО_Подготовка\\RealDataSet\\ссылки_новые.xlsx')
link_input = pd.read_csv("D:\\ProgPrj\\dsProjects\\ЦП_ЮФО_Подготовка\\RealDataSet\\dataset_maga.csv")
link_learn = pd.read_excel("D:\\ProgPrj\\dsProjects\\ЦП_ЮФО_Подготовка\\RealDataSet\\dataset_for_lr.xlsx")
```

```python

# Загрузка стоп-слов
nltk.download('stopwords')
stop_words = set(stopwords.words('russian'))

def clean_text(text):
    # Удаление специальных символов и лишних пробелов
    text = re.sub(r'[^\w\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text)

    # Приведение к нижнему регистру
    text = text.lower()

    # Токенизация
    words = text.split()

    # Удаление стоп-слов
    words = [word for word in words if word not in stop_words]

    # Сборка очищенного текста
    return ' '.join(words)
```

```python
link_input["Описание"] = link_input["Описание"].apply(clean_text)

link_learn["Технологии"] = link_learn["Технологии"].apply(clean_text)
link_learn["Профессия"] = link_learn["Профессия"].apply(clean_text)
```

```python
datasetTMP = pd.concat([link_input, link_learn["Технологии"], link_learn["Профессия"], link_new["Ссылка на продукт"]], axis=1)
```
